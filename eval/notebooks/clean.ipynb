```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/sales.csv')
```

```python
print(df.shape)
```

```python
df.dtypes
```

```python
# Calculate profit
df['profit'] = df['revenue'] - df['cost']
```

```python
# Summary by region
df.groupby('region').agg({
    'revenue': 'sum',
    'profit': 'sum'
}).reset_index().rename(columns={'revenue': 'total_revenue', 'profit': 'total_profit'})
```
